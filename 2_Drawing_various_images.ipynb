```python
#Drawing an image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

# Rectangle

img image
- pt1:vertex of rectangle
- pt2: vertex of rectangle opposite to pt1
- color:rectangle or brightness

```python
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,128),color=(0,255,0),thickness=5)
#pt1:top left
#pt2:bottom right
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
#pt1:top left
#pt2:bottom right
```

```python
plt.imshow(blank_img)
```

# Circles 

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(511,511),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

# Text

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
